```python
import xlsxwriter
from itertools import islice

newFile = xlsxwriter.Workbook('Manuel_utf8.xlsx')
hoja = newFile.add_worksheet()
datos = ['Fecha', 'Precipitacion', 'Evaporacion', 'TemperaturaMaxima', 'TemperaturaMinima']
for i in range(len(datos)):
    hoja.write(0, i, datos[i])
with open("Manuel_utf8.txt") as archivo:
    i = 1
    for linea in islice(archivo, 19, None):
        j = 0
        if len(linea) > 1:
            date = linea[:10]
            precp = linea[12:20]
            evap = linea[20:27]
            tmax = linea[27:34]
            tmin = linea[34:38]
            hoja.write(i,j,date)
            j += 1
            hoja.write(i,j,precp)
            j += 1
            hoja.write(i,j,evap)
            j += 1
            hoja.write(i,j,tmax)
            j += 1
            hoja.write(i,j,tmin)
            j += 1
            i += 1
newFile.close()
```
